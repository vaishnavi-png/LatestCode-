```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import joblib
import shap
import warnings
warnings.filterwarnings('ignore')
sns.set(style='whitegrid')

# Load dataset (adjust path if needed)
df = pd.read_csv('/content/adult.csv')   # change to your filename if different
print("Shape:", df.shape)
df.head()
```

```python
# Overview
df.info()

# Quick look at missing or weird values
df.isnull().sum()

for col in df.columns:
    cnt = df[col].astype(str).str.contains(r'\?', regex=True).sum()
    if cnt > 0:
        print(f"{col}: {cnt} missing")
```

```python
# Show exact column names (with repr to reveal spaces)
print("Columns (repr):")
for c in df.columns:
    print(repr(c))

# Quick info & sample missing patterns
print("\nInfo:")
print(df.info())
print("\nValue counts for 'workclass' top values:")
print(df['workclass'].value_counts(dropna=False).head(20))

# Show rows that contain '?' (typical placeholder in this dataset)
print("\nRows with '?' placeholders by column:")
for col in df.columns:
    if df[col].dtype == object:
        cnt = df[col].astype(str).str.contains('\\?').sum()
        if cnt>0:
            print(f"{col}: {cnt}")
```

**Normalize column names & strip whitespace**

```python
# Normalize column names: lower, strip, replace '.' with '_' and spaces with '_'
df.columns = [c.strip().lower().replace('.', '_').replace(' ', '_') for c in df.columns]
df.rename(columns={
    'marital_status':'marital_status'
}, inplace=True)

# Strip strings in object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()

print("Normalized columns:", list(df.columns))
```

**Replace placeholder missing values and drop / impute**

```python
# Replace common placeholder '?' with NaN
df.replace('?', np.nan, inplace=True)
df.replace('?', np.nan, inplace=True)  # double-run to be safe (some variants)

# Show counts of missing after replacement
print("Missing counts after replacing '?':")
print(df.isnull().sum())

# Strategy: drop rows where occupation or workclass or native_country are missing (critical categorical info)
df = df.dropna(subset=['workclass','occupation','native_country'])

# For any remaining less-critical missing categorical values, replace with 'Unknown'
for col in ['workclass','occupation','native_country','relationship','race','sex','marital_status','education']:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')

# For numeric columns (none expected to be missing usually), fill with median
for col in ['age','fnlwgt','education_num','capital_gain','capital_loss','hours_per_week']:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

print("After cleaning missing: shape =", df.shape)
print(df.isnull().sum())
```

```python
# Strip extra whitespace again (defensive)
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip()

# Drop duplicates
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]
print(f"Dropped duplicates: {before-after} rows removed. New shape: {df.shape}")
```

```python
# Binary income flag for modeling and easy plotting
df['income_flag'] = df['income'].apply(lambda x: 1 if '>50K' in str(x) else 0)

# Age group (binned) for grouped visuals
bins = [0, 25, 35, 45, 55, 65, 100]
labels = ['<=25','26-35','36-45','46-55','56-65','66+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Simplify marital.status categories
df['marital_status'] = df['marital_status'].replace({
    'Married-civ-spouse':'Married',
    'Married-spouse-absent':'Married',
    'Married-AF-spouse':'Married'
})

# Create a 'capital_net' feature as capital_gain - capital_loss (useful numeric summary)
df['capital_net'] = df['capital_gain'] - df['capital_loss']

# Flag long working hours (helpful for visuals)
df['long_hours_flag'] = (df['hours_per_week'] > 40).astype(int)

# Quick summary
df[['age','age_group','education','education_num','occupation','hours_per_week','income','income_flag']].head(5)
```

```python
# Cap extreme capital_gain / loss at 99th percentile for visualization stability
for col in ['capital_gain','capital_loss','capital_net','hours_per_week','fnlwgt']:
    if col in df.columns:
        upper = df[col].quantile(0.99)
        lower = df[col].quantile(0.01)
        df[col] = df[col].clip(lower, upper)

print("Numeric columns clipped to 1st-99th percentiles for stable visuals.")
```

# How do demographic, educational, and occupational factors influence income levels among adults?

# This specific visual focuses on an occupational factor, working hours per week.

Reveals whether longer working hours correlate with higher earnings.

why did I choose box plot? : Efficiently displays median, spread, and outliers for each income group, making it ideal to compare workload distribution between 50K and >50K earners.

Conclusion:
Higher earners generally work more hours per week, but overlap suggests job type and education also significantly impact income. **bold text**



```python
plt.figure(figsize=(8,5))
sns.boxplot(x='income_flag', y='hours_per_week', data=df)
plt.xticks([0,1], ['<=50K','>50K'])
plt.ylabel('Hours per week')
plt.title('Hours per week distribution by income class')
plt.savefig('hours_vs_income_box.png', dpi=300, bbox_inches='tight')
plt.show()
```

1) Contribution to problem statement:
Highlights how demographic, educational, and occupational numeric factors (like age, education level, hours worked, capital gain) correlate with income, revealing which variables most strongly influence earnings.

2) Why heatmap:
A heatmap efficiently visualizes multiple correlations at once, showing both strength and direction of relationships in a clear, compact grid.

 3. Conclusion:
Income is positively correlated with education level, capital gain, and working hours, while weakly linked with age; confirming that education and work-related factors have the greatest impact on income levels.
